```python
# This code block will transform all information from the excel file into a markdown text file
data = pd.read_excel('cleaned_van_gogh_letters.xlsx')

# Define a function to process each row into markdown format
def letter_to_markdown(row):
    # Base metadata
    markdown = f"### Letter {row['Letter']}\n"
    markdown += f"**From:** {row['From']}\n"
    markdown += f"**To:** {row['To']}\n"
    if pd.notna(row['Location']):
        markdown += f"**Location:** {row['Location']}\n"
    if pd.notna(row['Date']):
        markdown += f"**Date:** {row['Date']}\n"
    markdown += "\n"

    # Letter content
    markdown += "#### Original Text\n"
    markdown += f"> {row['Original Text']}\n\n" if pd.notna(row['Original Text']) else "> No original text available.\n\n"
    
    markdown += "#### Translation\n"
    markdown += f"> {row['Translation']}\n\n" if pd.notna(row['Translation']) else "> No translation available.\n\n"
    
    # Notes
    notes = []
    for i in range(1, 129):  
        note_col = f"Note {i}"
        if note_col in row and pd.notna(row[note_col]):
            notes.append(row[note_col])
    if notes:
        markdown += "#### Notes\n"
        for idx, note in enumerate(notes, start=1):
            markdown += f"{idx}. {note}\n"
    markdown += "\n"

    return markdown

# Apply the function to each row and save to a markdown file
with open("van_gogh_letters.md", "w", encoding="utf-8") as f:
    for _, row in data.iterrows():
        f.write(letter_to_markdown(row))
```
